```python
import win32com
import win32com.client
import subprocess
import time
import win32gui


 # ----Path del ejecutable de SAP
path = r"C:\Program Files (x86)\SAP\SAPGUI770\SAPgui\saplogon.exe"


subprocess.Popen(path)
hwnd = 0
start_time = time.time()
while not hwnd:
     hwnd = win32gui.FindWindow(None, 'SAP Logon 770')
     if time.time() - start_time > 30:
       pass   # Si se supera el tiempo máximo de espera, se sale de la función
     time.sleep(0.5)


SapGuiAuto = win32com.client.GetObject('SAPGUI')
application = SapGuiAuto.GetScriptingEngine

session = "- P20 Production ERP Logistics and Finance"
num_connections = 1

# Open multiple connections to the ERP
for i in range(num_connections):
    connection = application.OpenConnection(session, True)
    print("Connection", i+1, "opened")

# Create an empty dictionary to store the number of windows per connection
windows_per_connection = {}

# Loop through all connections
for i in range(application.Connections.Count):
    connection = application.Children(i)
    
    # Loop through all sessions in each connection
    total_windows = 0
    for j in range(connection.Children.Count):
        session = connection.Children(j)
        
        # Count the number of open windows in each session
        windows = session.Children
        num_windows = windows.Count
        
        # Add the number of windows in the session to the total for the connection
        total_windows += num_windows
    
    # Store the total number of windows for this connection in the dictionary
    windows_per_connection[connection.Name] = total_windows

# Print the total number of windows for each connection
for connection_name, num_windows in windows_per_connection.items():
    print("Number of open windows in connection", connection_name, ":", num_windows)
```
